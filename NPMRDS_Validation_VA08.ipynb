# Visualization of NPMRDS Data using Scatter Plots and Cumulative Distribution Plots

This document visualizes NPMRDS data using scatter plots and CDF diagrams. The main purpose of this is to xompare and validate NPMRDS data against Bluetooth Traffic Monitoring data and Probe based traffic data from Inrix Inc. This document does neither includes BTM nor Inrix data.
The time frame of data collection is from May 08, 2014 to May 18, 2014. The data includes both weekday traffic and weekend traffic. NPMRDS reports data in 5 minute aggregate bins. For each hour, the maximum number of data points is 60/5 = 12. Therefore for the period of study, that is 10 days, the maximum number of points per hour is 120. However, since no data is reported in NPMRDS if no vehicle is observed during a 5 minute period, the number of points per hour is far less than 120.

## Library Imports

The following cell imports all library packages necessary for the rest of the code.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
```

## User Defined Parameters

The following four lines take the directory of the data files, the file names, color order for the scatter plots, and hours to highlight in the CDF diagrams. Please use the comments adjacent to the lines as a guide to help with entering the correct values.

```python
main_path = 'C:/Projects/NPMRDS/' #This is the folder where NPMRDS data files are stored. Use UNIX style slashes
routes = ['VA08-0005', 'VA08-0012'] #Ensure that these are names of files. Should be a comma separated list. Ex: routes = ['VA08-0005', 'VA08-0012']
colors = ['green', 'blue' , 'red', 'orange', 'purple'] #Add or remove, reorder to your preference
hr_hilt = [14, 16, 17, 18] #The hours to highlight in the final CDF plots. Enter hours from 00..23, separated by commas. Ex: hr_hilt = [17, 18]
```

## Main Code Block

The following lines are the main body of the code. Please be careful editing below as it can break the code and cause errors.

```python
for corridor in routes:
    pth = main_path + corridor + '.csv'
    df = pd.read_csv(pth, na_values=['\\N'])
    df['date_time'] = pd.to_datetime(df['date_time'])
    tt = ['ttav', 'ttpc', 'ttft']

    tmc = df['tmc'].unique()
    mdf = df[df['tmc'] == tmc[0]].copy()
    tmc = tmc[1:]
    for T in tmc:
        dft = df[df['tmc'] == T].copy()
        mdf = pd.merge(mdf, dft, on='date_time', how='inner', sort=False, copy=True)
        for TT in tt:
            mdf[TT] = mdf[TT + '_x'] + mdf[TT + '_y']
        col_to_drop = ['ttav_x', 'ttpc_x', 'ttft_x', 'ttav_y', 'ttpc_y', 'ttft_y']
        mdf.drop(col_to_drop, 1, inplace=True)

    temp = pd.DatetimeIndex(mdf['date_time'])
    mdf['date'] = temp.date
    mdf['time'] = temp.time

    fig, axes = plt.subplots(figsize=(20,10), dpi=600)
    ncolor = 0
    for TT in tt:
        axes.plot(mdf['time'], (mdf[TT]/60), 'o', color=colors[ncolor], label=TT.upper())
        ncolor += 1
    axes.legend(loc=0)
    axes.set_xticks(['00:00:00', '02:00:00', '04:00:00', '06:00:00', '08:00:00', '10:00:00', '12:00:00', '14:00:00', '16:00:00', '18:00:00', '20:00:00', '22:00:00', '23:59:59'])
    axes.set_xlabel('Time')
    axes.set_ylabel('Travel Time (Minutes)')
    axes.set_title('{0} NPMRDS 24-hour Scatter Plot (Full Extent)'.format(corridor))

    fig, axes = plt.subplots(figsize=(20,10), dpi=600)
    ncolor = 0
    for TT in tt:
        axes.plot(mdf['time'], (mdf[TT]/60), 'o', color=colors[ncolor], label=TT.upper())
        ncolor += 1
    axes.legend(loc=0)
    axes.set_xticks(['00:00:00', '02:00:00', '04:00:00', '06:00:00', '08:00:00', '10:00:00', '12:00:00', '14:00:00', '16:00:00', '18:00:00', '20:00:00', '22:00:00', '23:59:59'])
    axes.set_ylim([0,15])
    axes.set_xlabel('Time')
    axes.set_ylabel('Travel Time (Minutes)')
    axes.set_title('{0} NPMRDS 24-hour Scatter Plot (BT Data Extent)'.format(corridor))

    percentiles = np.arange(0, 1.01, 0.01)
    for jst in hr_hilt:
        for TT in tt:
            fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10), dpi=600)
            for j in range(0, 24):
                sli = mdf[(mdf['time'] >= datetime.time(j, 0, 0, 0)) & (mdf['time'] <= datetime.time(j, 59, 59, 99))].copy()
                if len(sli[TT].dropna()) > 0:
                    d = sli[TT].describe(percentiles=percentiles)
                    if j == jst:
                        axes[0].plot(sli['time'], (sli[TT]/60), 'o', color='black')
                    else:
                        axes[0].plot(sli['time'], (sli[TT]/60), 'o', color='lightblue')
                    if j == jst:
                        axes[1].plot((d[4:105]/60), percentiles, '-', color='black', linewidth=2)
                    else:
                        axes[1].plot((d[4:105]/60), percentiles, '-', color='skyblue', linewidth=0.75)
            axes[0].set_ylim([0, 15])
            axes[0].set_xticks(['00:00:00', '03:00:00', '06:00:00', '09:00:00', '12:00:00', '15:00:00', '18:00:00', '21:00:00', '23:59:59'])
            axes[0].set_xlabel('Time')
            axes[0].set_ylabel('Travel Time (Minutes)')
            axes[0].set_title('{0} NPMRDS 24-hour Scatter Plot for {1} at {2}:00 (BT Data Extent)'.format(corridor, TT.upper(), jst))
            axes[1].set_ylim([0, 1])
            axes[1].set_xlim([0, 15])
            axes[1].set_yticks(np.arange(0, 1.05, 0.05))
            axes[1].set_xlabel('Travel Time (Minutes)')
            axes[1].set_ylabel('Cumulative Probability (Percentile)')
            axes[1].set_title('{0} NPMRDS Travel Time CDF Plot for {1} at {2}:00 (BT Data Extent)'.format(corridor, TT.upper(), jst))
```
